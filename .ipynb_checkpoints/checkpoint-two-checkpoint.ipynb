```python
# dropping the columns identified above so I can ignore some of the stats when I run .describe in the next cell
df = seattle_df.drop(['alert_text', 'confirmation_text','reg_end_date', 'age_as_of_date',
       'required_forms', 'Location 1', 'Location 2'], axis=1)
# run .shape to confirm that 7 columns were dropped (46-7=39)
df.shape

# copy to task 3 
```

```python
# check for duplicates
df_dupes = df[df.duplicated(keep=False)]

print(df_dupes)
# no duplicates 
```

```python
# describe for stats--mainly concerns with ages min/max, waitlisted, and 
df.describe()
```

```python
# This is a cleaning task, but since I have 2 different notebooks and the dataframes are isolated from one another, 
# I have to run all manipulations to the dataframe in both notebooks. 

# removing all rows that correspond to data for programs with a minimum age over 18 because this is strictly a youth program focus and the EDA visualizations are much too complicated with that unecessary information included

to_drop = df[df['min_age'] > 19].index #identifies the indices of that meet the condition and assigns them to "to_drop"
# to_drop.shape #this allows me to check that the right rows are being identified before I actually drop them and mess up my dataframe! results is 544 which is less than I'd assumed I would return
df.drop(to_drop, inplace=True) #this tells the main dataframe that this special gorup actually needs to be dropped now that they've been identified
# df.shape this confirmed that the rows dropped from the previous total of 4928

# copy to task 3 notebook
```

```python
# Need to get some numbers about the different activity titles: 
# Looking for unique Activity titles

df_activities = df['ActivityTitle'].nunique()
print(df_activities)
```

## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python
# Target ages for registrants of programs
# need to count the number of times each min_age appears and then use a line chart to show the spread of that.
# did a cleaning step to eliminate the ages over 18 so that data won't be part of the displayed information since it doesn't apply
# value counts will get the totals per individual age value from that column, and then using the sort_index will put it in ascending order so the chart is easier to read. 

ages = df.groupby(['min_age']).size()

# code for the line chart with ages on the x axis and the value counts on the y axis

plt.plot(ages.index, ages.values)

# labels
plt.xlabel('Minimum Age')
plt.ylabel('Number of Activities')
plt.title('Number of Activities by Minimum Age Requirement')





```

```python
# Top enrollments overall by brochure section title
# need to find the sum of registration per BrochureSectionTitle, so group BrochureSectionTitle and then sum the num_registered
categories = df.groupby('BrochureSectionTitle')['num_registered'].sum().reset_index()

categories = categories[categories['num_registered'] > 0]

# using a pie chart to show the breakdown of the different categories
plt.pie(categories['num_registered'], labels=categories['BrochureSectionTitle'], autopct='%1.1f%%')

# labels
plt.title('Number of Participants Registered per Brochure Section Title')


# the results below will allow me to clean up some of the unnecessary brochure sections before I continue my analysis. 
# For example the Childcare and Field trips categories are not necessary for the information I need to review. 
```
